🔹 Potential criteria for a worsening condition:

Feature	Threshold / Rule

Blood Glucose (BGL)	> 180 or < 70 → unstable

Systolic BP	> 140 (hypertension) or < 90 (hypotension)

Diastolic BP	> 90 or < 60

Heart Rate	> 100 or < 50 → abnormal

Body Temp	> 100.4°F (fever) or < 95°F (hypothermia)

SPO2	< 95% → possible respiratory distress

```python
# Define a function to flag worsening conditions
def is_condition_worsening(row):
    return (
        (row["Blood Glucose Level(BGL)"] > 180)
        or (row["Blood Glucose Level(BGL)"] < 70)
        or (row["Systolic Blood Pressure"] > 140)
        or (row["Systolic Blood Pressure"] < 90)
        or (row["Diastolic Blood Pressure"] > 90)
        or (row["Diastolic Blood Pressure"] < 60)
        or (row["Heart Rate"] > 100)
        or (row["Heart Rate"] < 50)
        or (row["Body Temperature"] > 100.4)
        or (row["Body Temperature"] < 95)
        or (row["SPO2"] < 95)
    )


# Apply to create the target column
df["Condition_Worsening"] = df.apply(is_condition_worsening, axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns="Condition_Worsening")
Y = df['Condition_Worsening']
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
x_train.shape , y_train.shape
```

```python
x_test.shape , y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)

x_test_scaled=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
model=LogisticRegression()
model.fit(x_train_scaled,y_train)
```

```python
y_pred=model.predict(x_test_scaled)
```

```python
# Get confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot confusion matrix
plt.figure(figsize=(5, 4))
sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=["Not Worsening", "Worsening"],
    yticklabels=["Not Worsening", "Worsening"],
)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.tight_layout()
plt.show()
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```
